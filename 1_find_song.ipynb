```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the webpage (Spotify Daily charts from 2018/03/14 to 2024/07/02.)
url = "https://kworb.net/spotify/country/vn_daily_totals.html"

# Send a GET request to fetch the HTML content
response = requests.get(url)
html_content = response.text

# Parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table containing the data
table = soup.find('table', class_='addpos')

# Initialize a list to store data
data = []

# Initialize a counter for numerical order
num_order = 1

# Iterate through each row in the table (skipping the header row)
for row in table.find_all('tr')[1:]:  # Skip header row
    # Extract columns (name, days, T10, Pk, (x?), PkStreams, Total)
    columns = row.find_all('td')

    if len(columns) >= 7:
        try:
            # Extract name and artist
            name_artist = columns[0].text.strip()
            name = name_artist.split(' - ')[1].strip()
            artist = name_artist.split(' - ')[0].strip()
            
            # Extract track URL
            track_url = columns[0].find_all('a')[1]['href']  # Second <a> tag is for track URL
            
            # Extract track ID from the URL correctly
            track_id = track_url.split('/')[-1].split('.')[0]

            # Append data as a tuple including numerical order
            data.append((num_order, name, artist, track_id))
            
            # Increment numerical order
            num_order += 1
        
        except IndexError as e:
            print(f"IndexError: {e} - Skipping row: {row}")
        
        except Exception as e:
            print(f"Error: {e} - Skipping row: {row}")

# Specify the CSV file path
csv_file = 'spotify_data_vn_all_ordered.csv'

# Write data to CSV file
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Order', 'Name', 'Artist', 'id'])  # Write header
    writer.writerows(data)

print(f"CSV file '{csv_file}' with ordered data has been created successfully.")
```
